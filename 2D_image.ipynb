```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#1.
f = open('실습_데이터_1106.txt', 'r')
header = f.readline()

x = []
y = []
rho = []

for line in f:
    line = line.strip()
    columns = line.split()
    x.append(float(columns[0]))
    y.append(float(columns[1]))
    rho.append(float(columns[3]))

f.close()

xdata = np.array(x)
ydata = np.array(y)
rho_log = np.log10(np.array(rho))
```

```python
plt.figure(figsize=(10,8))
plt.scatter(xdata, ydata, s=0.05, marker='o', c=rho_log, cmap='viridis')
cb = plt.colorbar()
cb.set_label('Log nh [$cm^(-3)$]')
plt.xticks([0, 1, 2, 3])
plt.yticks([0, 1, 2, 3])
plt.xlim(0,3.1)
plt.ylim(0,3.1)
plt.xlabel('x [comoving Mpc]')
plt.ylabel('y [comoving Mpc]')
```

```python
#2. 
n = 200

x = np.linspace(0, 3.1, n)
y = x

grid = np.zeros((n,n))
for i in range(n-1):
    ind = np.where((xdata > x[i]) & (xdata <= x[i+1]))
    for j in range(n-1):
        ind2 = np.where((ydata[ind] > y[j]) & (ydata[ind] <= y[j+1]))
        grid[i, j] = np.log10(len(ind2[0]))
        if (len(ind2[0]) == 0):
            grid[i, j] = 0

plt.figure(figsize=(10,8))
plt.imshow(grid, cmap = 'viridis', extent=(0, 3.1, 0, 3.1))
plt.xticks([0, 1, 2, 3])
plt.yticks([0, 1, 2, 3])
plt.xlabel('x [comoving Mpc]')
plt.ylabel('y [comoving Mpc]')
cb = plt.colorbar(shrink = 0.8)
cb.set_label('Log (# of stars)', fontsize=12)
ticklabs = ['0', '0.5', '1.0', '1.5', '2.0', '2.5', '3.0', '3.5']
cb.ax.set_yticklabels(ticklabs, fontsize=12)
```
